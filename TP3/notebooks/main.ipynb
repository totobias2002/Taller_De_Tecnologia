```python
import pandas as pd
import matplotlib.pyplot as gf1
import matplotlib.pyplot as gf2

data = pd.read_csv("../data/zapas.csv")



x = data['Cual_es_mi_tu_marca_de_ropa_fav'].value_counts().index.tolist()
y = data['Cual_es_mi_tu_marca_de_ropa_fav'].value_counts().tolist()

gf1.bar(x,y)
gf1.title('Marca de ropa fav.')
gf1.xlabel('Marcas')
gf1.ylabel('Cantidad de personas que le gusta esa marca')
gf1.xticks(rotation=90,fontsize=12)
gf1.show()

xx = data['Cual_de_estas_Zapatillas_te_gustan_mas'].value_counts().index.tolist()
yy = data['Cual_de_estas_Zapatillas_te_gustan_mas'].value_counts().tolist()

gf2.bar(xx,yy)
gf2.title('Modelo de Zapatillas')
gf2.xlabel('Zapatillas')
gf2.ylabel('Cantidad de personas que le gusta esas zapatillas')
gf2.xticks(rotation=90,fontsize=12)
gf2.show()
```
