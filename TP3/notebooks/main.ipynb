```python
import pandas as pd
import matplotlib.pyplot as gf1

data = pd.read_csv("../data/sextogrande.csv")



x = data['sexto_grande'].value_counts().index.tolist()
y = data['sexto_grande'].value_counts().tolist()

gf1.bar(x,y)
gf1.title('sexto grande')
gf1.xlabel('equipos')
gf1.ylabel('votos')
gf1.xticks(rotation=90,fontsize=12)
gf1.show()
```

llegue a la conclusion que la mayoria de la gente piensa que estudiantes de la plata es el sexto grande.
