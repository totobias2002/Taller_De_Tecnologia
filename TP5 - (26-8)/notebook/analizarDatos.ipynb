```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as grafico
df = pd.read_csv("../data/player_stats.csv")
```

#Preguntas
Cual es el jugador de CS:GO con mas mapas jugados?

```python
ordenado = df.sort_values(by=['total_maps'], ascending=False)
x = ordenado['name'].iloc[0:10]
y = ordenado['total_maps'].iloc[0:10]

grafico.bar(x,y)
grafico.xlabel('Jugador')
grafico.ylabel('Cantidad')
grafico.xticks(rotation=90,fontsize=12)
grafico.show()
```

Cual es el pais con mas jugadores?

```python
cantidad = df['country'].value_counts()
pais = df['country'].unique()
grafico.pie(cantidad, labels=pais, autopct='%1.1f%%')
grafico.show()
```

Cual es el pais con mejor promedio de KD?

```python
promedio = df.groupby('country')['kd'].mean().sort_values(ascending=False).iloc[0:10]
x = promedio.index
y = promedio

grafico.bar(x,y)
grafico.xlabel('Pais')
grafico.ylabel('Promedio')
grafico.xticks(rotation=90,fontsize=12)
grafico.show()
```

El promedio de KD representa mejor promedio de Rating?

```python
promedio = df.groupby('country')['rating'].mean().sort_values(ascending=False).iloc[0:10]
x = promedio.index
y = promedio

grafico.bar(x,y)
grafico.xlabel('Pais')
grafico.ylabel('Promedio')
grafico.xticks(rotation=90,fontsize=12)
grafico.show()
```

Cuales son los jugadores argentinos con mejor rating?

```python
dfArgentinos = df[df['country'] == 'Argentina']

x = dfArgentinos['name']
y = dfArgentinos['rating']

grafico.bar(x,y)
grafico.xlabel('Nombre')
grafico.ylabel('Promedio')
grafico.xticks(rotation=90,fontsize=12)
grafico.show()
```
