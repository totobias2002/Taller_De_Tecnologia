```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/2019-20-fullyr-data_sa_crime.csv")
df.drop_duplicates()
df.dropna(inplace=True)
```

```python
def dejarPrimeros4ElementosLista(list):
    i=0;
    while i < len(list):
        if(i>4):
             list[i] = ""
        i = i+1
    return list

fecha = "01/07/2019"
print(fecha[3:5])

def returnMes(fecha):
    fecha = str(fecha)
    mes = fecha[3:5]
    if mes == "07":
        return "Julio"
    elif mes == "08":
        return "Agosto"
    elif mes == "09":
        return "Septiembre"
    elif mes == "10":
        return "Octubre"
    elif mes == "11":
        return "Noviembre"
    elif mes == "12":
        return "Diciembre"
    elif mes == "01":
        return "Enero"
    elif mes == "02":
        return "Febrero"
    elif mes == "03":
        return "Marzo"
    elif mes == "04":
        return "Abril"
    elif mes == "05":
        return "Mayo"
    elif mes == "06":
        return "Junio"
```

# Preguntas
1. ¿Cuáles son las ofensas más frecuentes?(Descripción nivel 2)

```python
import matplotlib.pyplot as plt
x_values = df['Offence Level 2 Description'].value_counts().index.tolist()
y_values = df['Offence Level 2 Description'].value_counts().tolist()

x_values = dejarPrimeros4ElementosLista(x_values)

plt.pie(y_values, labels=x_values)
plt.show()
```

2. ¿Cuál es la ofensa más frecuente?(Descripción nivel 3)

```python
import matplotlib.pyplot as plt
x_values = df['Offence Level 3 Description'].value_counts().index.tolist()
y_values = df['Offence Level 3 Description'].value_counts().tolist()


fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(x_values,y_values)
plt.xticks(rotation = 90)
plt.show()
```

3. ¿Hay más ofensas hacia propiedad o personas?

```python
import matplotlib.pyplot as plt
x_values = df['Offence Level 1 Description'].value_counts().index.tolist()
y_values = df['Offence Level 1 Description'].value_counts().tolist()

plt.pie(y_values, labels=x_values)
plt.show()
```

4. ¿En qué meses hubo más ofensas?

```python
dfConMes = df
dfConMes["Reported Date"] = dfConMes["Reported Date"].apply(returnMes)

x_values = dfConMes['Reported Date'].value_counts().index.tolist()
y_values = dfConMes['Reported Date'].value_counts().tolist()

plt.pie(y_values, labels=x_values)
plt.show()
```

5. ¿En qué localidades hubo más ofensas?

```python
import matplotlib.pyplot as plt
x_values = df['Suburb - Incident'].value_counts().index.tolist()
y_values = df['Suburb - Incident'].value_counts().tolist()

x_values = dejarPrimeros4ElementosLista(x_values)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(x_values,y_values)
plt.xticks(rotation = 90)
plt.show()
```
