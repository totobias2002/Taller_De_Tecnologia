```python
import pandas as pd
import numpy as np
import seaborn as sns

df = pd.read_csv("../data/historico_temperaturas.csv")
df.drop_duplicates()
df.dropna(inplace=True)

df1 = df[["maxima", "mes", "anio" ]]
grafico = sns.pairplot(df1,hue="mes",palette="Spectral")
```

```python
df2 = df[["minima", "mes", "anio" ]]
grafico = sns.pairplot(df2,hue="mes",palette="Spectral")
```

```python
df3 = df[["media", "mes", "anio" ]]
sns.pairplot(df3,hue="mes",palette="Spectral")
```
