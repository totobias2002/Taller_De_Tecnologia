```python
import pandas as pd
import matplotlib.pyplot as gf1
import matplotlib.pyplot as gf2
import numpy as np
data=pd.read_csv("../data/solicitudes-acceso-informacion-publica-20210712.csv")
data=data.dropna()
data=data.loc[0:500]
```

```python
x=data['area_emisora'].unique()
y=data['area_emisora'].value_counts().tolist()
gf1.bar(x,y)
gf1.title('areas emisoras')
gf1.xlabel('areas')
gf1.ylabel('cantidad de areas')
gf1.xticks(rotation=90,fontsize=12)
gf1.show()
```

```python
x2=data['genero_solicitante'].unique()
y2=data['genero_solicitante'].value_counts().tolist()
gf2.plot(x2,y2)
gf2.title('generos')
gf2.xlabel('genero')
gf2.ylabel('tipo de generos')
gf2.xticks(rotation=90,fontsize=12)
gf2.show()
```
