Se realizo un formulario para analizar como fue el rendimiento de
los alumnos con respecto a la materia de Taller de Soporte Informatico,
y estos fueron los resultados

```python
import pandas as pd
import matplotlib.pyplot as gr
import matplotlib.pyplot as gr2
import matplotlib.pyplot as gr3
import matplotlib.pyplot as gr4
import matplotlib.pyplot as gr5
import matplotlib.pyplot as gr6
import matplotlib.pyplot as gr7
import matplotlib.pyplot as gr8
import matplotlib.pyplot as gr9
import matplotlib.pyplot as gr10

data = pd.read_csv("../data/analisisTallerDeSoporteInformático.csv")
```

```python
gr.title('¿Pudo entregar todos los trabajos de la materia?')
cant = [data["¿Pudo entregar todos los trabajos de la materia?"].value_counts()["Sí"],data["¿Pudo entregar todos los trabajos de la materia?"].value_counts()["No"]]
sino = ["Sí","No"]
gr.pie(cant, labels=sino, autopct='%1.1f%%')
gr.show()
```

```python
x = data['¿Cuánto te gusta Python como lenguaje de programación?'].unique()
y = data['¿Cuánto te gusta Python como lenguaje de programación?'].value_counts().tolist()
gr2.bar(x,y)
gr2.xlabel('Gusto')
gr2.ylabel('Cant de alumnos')
gr2.xticks(rotation=90,fontsize=11)
gr2.show()
```

```python
gr3.title('¿Consideras que tu rendimiento por la mañana es peor que por la tarde?')
cant = [data["¿Consideras que tu rendimiento por la mañana es peor que por la tarde?"].value_counts()["Sí"],data["¿Consideras que tu rendimiento por la mañana es peor que por la tarde?"].value_counts()["No"]]
sino = ["Sí","No"]
gr3.pie(cant, labels=sino, autopct='%1.1f%%')
gr3.show()
```

```python
gr4.title('¿Considerás que analizar datos es aburrido?')
cant = [data["¿Considerás que analizar datos es aburrido?"].value_counts()["Sí"],data["¿Considerás que analizar datos es aburrido?"].value_counts()["No"]]
sino = ["Sí","No"]
gr4.pie(cant, labels=sino, autopct='%1.1f%%')
gr4.show()
```

```python
gr5.title('¿Tu cumplimiento con las entregas tiene que ver con que estás en el último año y querés terminar?')
cant = [data["¿Tu cumplimiento con las entregas tiene que ver con que estás en el último año y querés terminar?"].value_counts()["Sí"],data["¿Tu cumplimiento con las entregas tiene que ver con que estás en el último año y querés terminar?"].value_counts()["No"]]
sino = ["Sí","No"]
gr5.pie(cant, labels=sino, autopct='%1.1f%%')
gr5.show()
```

```python
gr6.title('¿Se te dificultó la búsqueda de datasets? ')
cant = [data["¿Se te dificultó la búsqueda de datasets? "].value_counts()["Sí"],data["¿Se te dificultó la búsqueda de datasets? "].value_counts()["No"]]
sino = ["Sí","No"]
gr6.pie(cant, labels=sino, autopct='%1.1f%%')
gr6.show()
```

```python
gr7.title('¿Te resultó GitHub una herramienta útil para ver los enunciados de la materia?')
cant = [data["¿Te resultó GitHub una herramienta útil para ver los enunciados de la materia? "].value_counts()["Sí"],data["¿Te resultó GitHub una herramienta útil para ver los enunciados de la materia? "].value_counts()["No"]]
sino = ["Sí","No"]
gr7.pie(cant, labels=sino, autopct='%1.1f%%')
gr7.show()
```

```python
x = data['¿Crees que el cambio de modalidad (Virtual y presencial) dificultó tu rendimiento?'].unique()
y = data['¿Crees que el cambio de modalidad (Virtual y presencial) dificultó tu rendimiento?'].value_counts().tolist()
gr8.bar(x,y)
gr8.xlabel('Si/No')
gr8.ylabel('Cant de alumnos')
gr8.xticks(rotation=90,fontsize=11)
gr8.show()
```

```python
x = data['¿Hasta qué mes entregaste trabajos?'].unique()
y = data['¿Hasta qué mes entregaste trabajos?'].value_counts().tolist()
gr10.bar(x,y)
gr10.xlabel('Meses')
gr10.ylabel('Cant de alumnos')
gr10.xticks(rotation=90,fontsize=11)
gr10.show()
```
