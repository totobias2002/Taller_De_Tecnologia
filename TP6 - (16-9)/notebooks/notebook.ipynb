```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

if __name__ == '__main__':
    headers = {"User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36"}
    pagina = requests.get("https://www.metacritic.com/person/lana-del-rey", headers=headers)
    print("codigo ", pagina.status_code)
    objeto_sopa = BeautifulSoup(pagina.content, 'html.parser')
    #print(objeto_sopa.prettify())
    listaDeObjetos = list(objeto_sopa.children)

    tbody = objeto_sopa.find('tbody')
    print(tbody)
    datos = {}

    e1 = list(tbody.find_all('tr'))

    listaMetascore = list()
    listaTitulo = list()
    listaRole = list()
    listaScore = list()

    for i in range(len(e1)):

        listaMetascore.insert(i, e1[i].find('span', class_='metascore_w').getText())
        print(i, e1[i].find('span', class_='metascore_w').getText())
        listaTitulo.insert(i, e1[i].find('a').getText())
        print(i, e1[i].find('a').getText())
        listaRole.insert(i, e1[i].find('td', class_="role").getText())
        print(i, e1[i].find('td', class_="role").getText())
        listaScore.insert(i, e1[i].find('td', class_="score").find('span').getText())
        print(i, e1[i].find('td', class_="score").find('span').getText())

    dataframe = pd.DataFrame(list(zip(listaMetascore, listaTitulo, listaRole, listaScore)),columns=['Metascore','Titulo', 'Rol', 'Puntuación de usuario'])
    print(dataframe)
    dataframe.to_csv("../data/LanaMetacritic.csv")
```
