```python


import pandas as pd
import numpy as np
import seaborn as sns

df = pd.read_csv("../data/2006_-_2011_NYS_Math_Test_Results_by_Grade_-_Citywide_-_by_Race-Ethnicity.csv")
df.drop_duplicates()
df.dropna(inplace=True)

df1 = df[["Year", "Number Tested", "Mean Scale Score", ]]
grafico = sns.pairplot(df1,hue="Year",palette="Spectral")
```

1. A medida que pasan los años, hubieron más testeados (exceptuando el 2011)
2. Aumentan las calificaciones a lo largo de los años


```python
df2 = df[["Category", "Mean Scale Score", "Year", "Number Tested"]]
grafico = sns.pairplot(df2,hue="Category",palette="icefire")
```

3. Los asiaticos suelen tener mas calificaciones que el resto, despues siguen los blancos y abajo están los negros y latinos.
